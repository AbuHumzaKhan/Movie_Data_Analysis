```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('TMDB_movie_dataset_v11.csv')
df.head()
```

```python
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(' ', '_')
              .str.replace(r'[^a-zA-Z0-9_]', '', regex=True)
)
print("Columns after cleaning:\n", df.columns.tolist())
```

```python
# 2. Remove duplicates
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

```python
# 3. Remove columns with only one value

for col in df.columns:
    if df[col].nunique() <= 1:
        df = df.drop(columns=[col])
```

```python
# 4. Handle missing values
# Droping columns with more than 80% missing
missing = df.isnull().mean() * 100
high_missing = missing[missing > 80].index
df = df.drop(columns=high_missing)
```

```python
# Fill numbers with median
for col in df.select_dtypes(include="number").columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Fill text with mode
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# 5. Convert dates (if column name has "date")
for col in df.columns:
    if "date" in col.lower():
        df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
# Missing values plot
missing = df.isnull().mean() * 100
missing = missing[missing > 0]
if not missing.empty:
    plt.figure(figsize=(8,4))
    sns.barplot(x=missing.values, y=missing.index)
    plt.title("Missing Values %")
    plt.show()
```

```python
# Histogram for first 3 numeric columns
for col in df.select_dtypes(include="number").columns[:3]:
    plt.figure()
    sns.histplot(df[col], bins=20, kde=True)
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
# Convert release_date to datetime
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")

# Extract year
df["release_year"] = df["release_date"].dt.year

# Find year with most releases
year_counts = df["release_year"].value_counts().sort_values(ascending=False)

print("ðŸŽ¬ Year with the most movies released:")
print(year_counts.head(1))
```

```python
oldest = df.loc[df["release_date"].idxmin()]
```

```python
# Drop rows where revenue is missing or 0
df_revenue = df[df["revenue"] > 0]

# Find highest revenue movie
highest = df_revenue.loc[df_revenue["revenue"].idxmax()]

print("ðŸ’° Highest Grossing Movie:")
print(highest[["title", "release_year", "revenue"]])
```

```python


# Convert release_date to datetime and extract year
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")
df["release_year"] = df["release_date"].dt.year

# Filter only movies with revenue > 0
df_revenue = df[df["revenue"] > 0]

# Top 5 movies by revenue
top5 = df_revenue.sort_values("revenue", ascending=False).head(5)

print("ðŸ’° Top 5 Movies by Revenue:")
print(top5[["title", "release_year", "revenue"]])
```

```python
# Create profit column
df["profit"] = df["revenue"] - df["budget"]
```

```python
# 1. How many movies are there in total?
print("Total movies:", df.shape[0])
```

```python
# 2. Distribution of movies over the years
plt.figure(figsize=(10,6))
df["release_year"].value_counts().sort_index().plot(kind="bar")
plt.title("Number of Movies Released per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Which year had the highest number of movie releases?
top_year = df["release_year"].value_counts().idxmax()
print("Year with most movie releases:", top_year)
```

```python
# 5. How many movies are marked as 'adult'?
print("Adult movies count:", df["adult"].sum())
```

```python
# 6. Top 10 movies by revenue
print(df.sort_values("revenue", ascending=False)[["title", "revenue"]].head(10))
```

```python
# 7. Top 10 movies by profit
print(df.sort_values("profit", ascending=False)[["title", "profit"]].head(10))
```

```python
# 8. Average budget vs revenue over time
avg_budget_revenue = df.groupby("release_year")[["budget", "revenue"]].mean()
avg_budget_revenue.plot(figsize=(10,6))
plt.title("Average Budget vs Average Revenue per Year")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.show()
```

```python
# 9. Which year had the most profitable movies?
year_profit = df.groupby("release_year")["profit"].sum()
print("Most profitable year:", year_profit.idxmax())
```

```python
# 10. Trend of global movie revenues
year_revenue = df.groupby("release_year")["revenue"].sum()
year_revenue.plot(figsize=(10,6))
plt.title("Total Revenue per Year")
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.show()
```
